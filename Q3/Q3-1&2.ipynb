```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams.update({
    "font.family": "Times New Roman",
    "axes.titlesize": 22,
    "axes.labelsize": 18,
    "xtick.labelsize": 14,
    "ytick.labelsize": 14,
})

palette_top10 = [
    "#5975a4", "#cc8963", "#5f9e6e", "#857aab", "#8d7866",
    "#d095bf", "#c1b37f", "#71aec0", "#402141", "#832b5a"
]

top8_palette = [
    "#5B7DB1",  # 
    "#7F9CC8",
    "#A9BBD9",
    "#D9DDE5",  # 
    "#E6C3BD",
    "#D59A92",
    "#C46E68",
    "#A84A46"   # 
]

top10_dancers = pd.Series(
    [6.0, 5.9, 5.8, 5.7, 5.6, 5.5, 5.4, 5.2, 4.1, 3.0],
    index=[
        "Sasha Farber",
        "Maksim Chmerkovskiy",
        "Cheryl Burke",
        "Witney Carson",
        "Lindsay Arnold",
        "Valentin Chmerkovskiy",
        "Mark Ballas",
        "Daniella Karagach",
        "Julianne Hough",
        "Derek Hough"
    ]
)

top8_industry = pd.Series(
    [5.0, 3.0, 3.0, 3.0, 2.0, 2.0, 2.0, 1.0],
    index=[
        "Racing Driver",
        "Inspirational Speaker",
        "Military",
        "Fitness Instructor",
        "Digital Creator",
        "Producer",
        "Musician",
        "Conservationist"
    ]
)

def plot_bar_vertical(s, title, xlabel, ylabel, out_png, colors, rotate_deg=30):
    """
    s: pd.Series, index=, values=
    colors: list[str],  >= len(s)
    """
    s = s.sort_values(ascending=False)
    n = len(s)
    use_colors = colors[:n]

    fig, ax = plt.subplots(figsize=(12, 5), dpi=200)
    ax.bar(s.index.astype(str), s.values, color=use_colors, edgecolor="none")

    ax.set_title(title, pad=10)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)

    ax.set_xticklabels(ax.get_xticklabels(), rotation=rotate_deg, ha="right")

    sns.despine(ax=ax, top=True, right=True)
    ax.margins(y=0.08)

    plt.tight_layout()
    plt.savefig(out_png, bbox_inches="tight")
    plt.show()

plot_bar_vertical(
    top10_dancers,
    title="Top10 Professional Dancers' Coaching Effect",
    xlabel="Professional Dancer",
    ylabel="Average Score",
    out_png="top10_dancers_vertical_unique_colors.png",
    colors=palette_top10,
    rotate_deg=35
)

plot_bar_vertical(
    top8_industry,
    title="Different Industry Performance",
    xlabel="Industry",
    ylabel="Average Score",
    out_png="top8_industry_vertical_diverging_colors.png",
    colors=top8_palette,
    rotate_deg=25
)
```
